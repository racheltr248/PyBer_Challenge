### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

```python
pyber_data_df.to_csv('./Resources/pyber_data.csv')
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# use groupby() and count() on "ride_id" column
total_rides_by_type = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_by_type
```

```python
# 2. Get the total drivers for each city type
# use groupby() and sum() on "driver_count" column
total_drivers_by_type = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
# use groupby() and sum() on the "fare" column
total_fares_by_type = pyber_data_df.groupby(['type']).sum()['fare']
total_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
# calculate the average fare per ride by dividing the sum of all fares by total rides
average_fare_per_ride = total_fares_by_type / total_rides_by_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
# calculate the average fare per driver by dividing the sum of all fares by total drivers
average_fare_per_driver = total_fares_by_type / total_drivers_by_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fares_by_type,
    "Average Fare per Ride": average_fare_per_ride, 
    "Average Fare per Driver": average_fare_per_driver
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
# , $ 2 decimals
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Read the merged DataFrame
pyber_data_df
```

```python
# 1. Show the total fare amount for each date
# use groupby() on "type" and "date", then sum on "fare"
sum_fares_by_type = pyber_data_df.groupby(['type', 'date']).sum()['fare']
sum_fares_by_type
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# use following code snippet
sum_fares_by_type_df = sum_fares_by_type.reset_index()
sum_fares_by_type_df
```

```python
# 3. Create a pivot table to get the total fares for each type of city by the date.
# use pivot() so index is "date", columns are each a city "type", and values are "fare"
sum_fares_by_type_pivot = sum_fares_by_type_df.pivot(index='date', columns='type', values='fare')
sum_fares_by_type_pivot
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc 
# use loc() on date range 2019-01-01 through 2019-04-28
fares_jan_apr = sum_fares_by_type_pivot.loc['2019-01-01':'2019-04-30']
fares_jan_apr
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# use following code snippet
fares_jan_apr.index = pd.to_datetime(fares_jan_apr.index)
fares_jan_apr
```

```python
# 6. Check that the datatype for the index is datetime 
# use following code snippet
fares_jan_apr.info()
```

```python
# 7. Create a new DataFrame and get the total fares for each week
# use resample() by week 'W', sum total fares for each week
# confirm that dataframe looks like fare_per_week image in resources
weekly_fares_df = fares_jan_apr.resample('w').sum()
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# confirm that graph looks like Challenge_fare_summary image in resources

ax = weekly_fares_df.plot(figsize = (20, 5))

# title
ax.set_title('Total Fare by City Type')

# ylabel
ax.set_ylabel('Fare ($USD)')

# legend
ax.legend(loc='center')

# save figure

plt.savefig('./Analysis/Fare_Summary_Plot.png')
plt.show()
```
